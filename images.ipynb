```python
from PIL import Image
import matplotlib.pyplot as plt
import cv2
import numpy as np
image = cv2.imread('/Users/eddie/Downloads/No Gas (1)/No Gas/121.png')
image_norm = cv2.resize(image,(224,224))
image_norm = image_norm/255.0
sharpen_kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]])
sharpen = cv2.filter2D(image_norm, -1, sharpen_kernel)
imgplot = cv2.imshow('sharpen',sharpen)
imgplot2 = cv2.imshow('original',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
